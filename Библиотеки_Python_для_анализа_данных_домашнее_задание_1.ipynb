Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
filtered_movie = ratings[(ratings['rating'] == 5.0)]
filtered_movie.head()
```

```python
joined = filtered_movie.merge(movies, on = 'movieId', how = 'left')
joined.head()
```

```python
joined.value_counts('movieId', ascending = False).head(1)
```

```python
result = joined[joined['movieId'] == 318]

result
```

Не учитывайте в расчётах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power.head()
```

```python
power[power['country'].str.contains('lat', case = False)]['country'].unique()
```

```python
power[power['country'].str.contains('lit', case = False)]['country'].unique()
```

```python
power[power['country'].str.contains('est', case = False)]['country'].unique()
```

```python
tab = power[
    (
        (power['country'].isin(['Latvia', 'Lithuania', 'Estonia']))
        &
        (power['category'].isin([4, 12, 21]))
        &
        (power['year'].isin([2005, 2006, 2007, 2008, 2009, 2010]))
        &
        (power['quantity'] >= 0)
    )
]
tab.head()
```

```python
tab['quantity'].sum()
```
