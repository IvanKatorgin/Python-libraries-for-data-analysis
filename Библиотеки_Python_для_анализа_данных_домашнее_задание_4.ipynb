Задание 1. Обязательная часть

Вам необходимо написать функцию, которая будет основана на поиске по сайту habr.com (https://habr.com/ru/search/).

Функция в качестве параметра должна принимать список запросов для поиска (например, ['python', 'анализ данных']) и на основе материалов, попавших в результаты поиска по каждому запросу, возвращать датафрейм вида:

<дата> - <заголовок> - <ссылка на материал>

В рамках задания предполагается работа только с одной (первой) страницей результатов поисковой выдачи для каждого запроса. Материалы в датафрейме не должны дублироваться, если они попадали в результаты поиска для нескольких запросов из списка.

```python
import pandas as pd, requests, time
from bs4 import BeautifulSoup

def get_articles(queries):
  article = pd.DataFrame()
  url = 'https://habr.com/ru/search/'
  for query in queries:
    params = {'q': query}
    res = requests.get(url, params)
    soup = BeautifulSoup(res.text)
    refs = soup.find_all('article', class_='tm-articles-list__item')
    for ref in refs:
      date = ref.find('a', class_ = 'tm-article-datetime-published').text
      title = ref.find('a', class_ = 'tm-title__link').text
      link = 'https://habr.com' + ref.find('a', class_='tm-title__link').get('href')
      row = {'дата': date, 'заголовок': title, 'ссылка на материал': link}
      article = pd.concat([article, pd.DataFrame([row])])
  return article.reset_index(drop=True).drop_duplicates(subset = ['заголовок', 'ссылка на материал'])

get_articles(['python', 'анализ данных'])
```

Задание 2. Дополнительная часть (необязательная)

Функция из обязательной части задания должна быть расширена следующим образом:

 - кроме списка ключевых слов для поиска необходимо объявить параметр с количеством страниц поисковой выдачи. Т.е. при передаче в функцию аргумента 4 необходимо получить материалы с первых 4 страниц результатов;

 - в датафрейме должны быть столбцы с полным текстом найденных материалов и количеством лайков:

  <дата> - <заголовок> - <ссылка на материал> - <текст материала> - <количество лайков>

```python
import pandas as pd, requests, time
from bs4 import BeautifulSoup

def get_articles(queries, pages):
  article = pd.DataFrame()
  url = 'https://habr.com/ru/search/'
  for query in queries:
    params = {'q': query}
    for page in range (1, pages+1):
      params['page'] = url
      res = requests.get(url, params)
      soup = BeautifulSoup(res.text)
      time.sleep(0.3)
      refs = soup.find_all('article', class_='tm-articles-list__item')
      for ref in refs:
        date = ref.find('a', class_ = 'tm-article-datetime-published').text
        title = ref.find('a', class_ = 'tm-title__link').text
        link = 'https://habr.com' + ref.find('a', class_='tm-title__link').get('href')
        text_article = requests.get(link)
        time.sleep(0.2)
        soup_info = BeautifulSoup(text_article.text)
        article_body = soup_info.find('p').text
        likes = ref.find('svg', class_ = 'tm-svg-img tm-votes-meter__icon tm-votes-meter__icon tm-votes-meter__icon_appearance-article').text
        raitings = ref.find('span', class_ = 'tm-votes-meter__value').text
        row = {'дата': date, 'заголовок': title, 'ссылка на материал': link, 'текст материала': article_body, 'рейтинг': raitings, 'количество лайков': likes}
        article = pd.concat([article, pd.DataFrame([row])])
  return article.reset_index(drop=True).drop_duplicates(subset = ['заголовок', 'ссылка на материал'])

get_articles(['python', 'анализ данных'], 4)
```
